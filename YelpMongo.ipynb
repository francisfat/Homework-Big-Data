##### 1) Query User Activity Over The Years (using date)

```python
from pymongo import MongoClient

# Requires the PyMongo package.
# https://api.mongodb.com/python/current

client = MongoClient('mongodb+srv://francisfat:varchetella@cluster0.c8bwx.mongodb.net/admin?retryWrites=true&w=majority')
result = client['yelp']['tips'].aggregate([
    {
        '$project': {
            'timestamp': {
                '$split': [
                    '$date', ' '
                ]
            }
        }
    }, {
        '$set': {
            'date': {
                '$arrayElemAt': [
                    '$timestamp', 0
                ]
            }, 
            'hour': {
                '$arrayElemAt': [
                    '$timestamp', 1
                ]
            }
        }
    }, {
        '$project': {
            'date': {
                '$split': [
                    '$date', '-'
                ]
            }
        }
    }, {
        '$set': {
            'year': {
                '$arrayElemAt': [
                    '$date', 0
                ]
            }, 
            'month': {
                '$arrayElemAt': [
                    '$date', 1
                ]
            }, 
            'day': {
                '$arrayElemAt': [
                    '$date', 2
                ]
            }
        }
    }, {
        '$group': {
            '_id': '$year', 
            'Number of Tips': {
                '$count': {}
            }
        }
    }, {
        '$sort': {
            'Number of Tips': -1
        }
    }
])
```

##### 3) Query About Dirtiest Places For Users (join & sent-an)

```python
import re
from pymongo import MongoClient

# Requires the PyMongo package.
# https://api.mongodb.com/python/current

client = MongoClient('mongodb+srv://francisfat:varchetella@cluster0.c8bwx.mongodb.net/admin?retryWrites=true&w=majority')
result = client['yelp']['tips'].aggregate([
    {
        '$match': {
            '$or': [
                {
                    'text': re.compile(r"dirly")
                }, {
                    'text': re.compile(r"bad smell")
                }, {
                    'text': re.compile(r"dingy")
                }, {
                    'text': re.compile(r"filthy")
                }, {
                    'text': re.compile(r"nasty")
                }
            ]
        }
    }, {
        '$project': {
            '_id': 1, 
            'business_id': 1, 
            'text': 1
        }
    }, {
        '$lookup': {
            'from': 'business', 
            'localField': 'business_id', 
            'foreignField': 'business_id', 
            'as': 'business'
        }
    }, {
        '$replaceRoot': {
            'newRoot': {
                '$mergeObjects': [
                    {
                        '$arrayElemAt': [
                            '$business', 0
                        ]
                    }, '$$ROOT'
                ]
            }
        }
    }, {
        '$project': {
            'business_id': 1, 
            'name': 1
        }
    }, {
        '$limit': 500
    }, {
        '$group': {
            '_id': '$business_id', 
            'Place Name': {
                '$first': '$name'
            }, 
            'Number of Frightened Users': {
                '$count': {}
            }
        }
    }, {
        '$sort': {
            'Number of Frightened Users': -1
        }
    }
])
```

##### 4) Query About Cities With More Sushi Restaurant

```python
import re
from pymongo import MongoClient

# Requires the PyMongo package.
# https://api.mongodb.com/python/current

client = MongoClient('mongodb+srv://francisfat:varchetella@cluster0.c8bwx.mongodb.net/admin?retryWrites=true&w=majority')
result = client['yelp']['business'].aggregate([
    {
        '$match': {
            'stars': {
                '$gt': 4
            }, 
            'categories': {
                '$regex': re.compile(r"sushi(?i)")
            }
        }
    }, {
        '$group': {
            '_id': '$city', 
            'State': {
                '$first': '$state'
            }, 
            'Total Sushi Restaurant': {
                '$count': {}
            }
        }
    }, {
        '$sort': {
            'Total Sushi Restaurant': -1
        }
    }
])
```
